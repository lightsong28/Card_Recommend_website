# 1. 이전 데이터 생성(미리 사이트에 저장된 고객 정보)

```python
# 데이터 생성에서 기준이 되는 금액 및 카테고리 대분류를 확인
customer_info = pd.read_csv('./customer_info.csv',encoding='cp949')
industry_code = pd.read_csv('./industry_code.csv',encoding='cp949')

def change_UPJONG_CD(x):
  return  str(x[:2].upper())+str(x[2:])
industry_code['업종코드(UPJONG_CD)'] = industry_code['업종코드(UPJONG_CD)'].map(change_UPJONG_CD)

# merge를 통해 업종코드별 대분류,중분류,소분류 열을 표시
age_ujcode = pd.merge(customer_info,industry_code,left_on = '서울시민업종코드(UPJONG_CD)', right_on = '업종코드(UPJONG_CD)')
# 카테고리 대분류 리스트를 class_list에 저장
category_list = age_ujcode['대분류(CLASS1)'].unique()
category_list

min_max_dict = {}

for i in category_list:
  min_max_list = []

  # 카테고리에 해당하는 이용금액, 건수 열만 저장
  df_buffer = age_ujcode[age_ujcode['대분류(CLASS1)']==i][['카드이용금액계(AMT_CORR)','카드이용건수계(USECT_CORR)']]
  
	# 카테고리별 하루 평균 사용 금액[(이용금액/이용건수)/30]의 최소, 최대를 mins,maxs에 저장
  mins = ((df_buffer['카드이용금액계(AMT_CORR)']/df_buffer['카드이용건수계(USECT_CORR)'])/30).min()
  if mins < 1000:
    mins = 1000
  maxs = ((df_buffer['카드이용금액계(AMT_CORR)']/df_buffer['카드이용건수계(USECT_CORR)'])/30).max()
  
	# 소수점 첫째자리에서 반올림하고 mins,maxs 리스트에 append후 dict value로 저장  
  min_max_list.append(round(mins))
  min_max_list.append(round(maxs))
  min_max_dict[i] = min_max_list
```

```python
import re

for i in range(len(card100_custom)):
#3. 연회비중 해외 겸용 금액만을 저장
  fee = card100_custom.loc[i,'연회비']
  regex = re.compile('\d*,\d*')
  annual_fee = regex.findall(fee)
  card100_custom.loc[i,'연회비'] = annual_fee[-1].replace(',','')

#4. 전월실적중 있는 것은 금액을 적고, 없는것은 0으로
  previous_m = card100_custom.loc[i,'전월실적']
  if '없음' in previous_m:
    card100_custom.loc[i,'전월실적'] = 0
  else:
    regex = re.compile(r'\d\d')
    matchobj = regex.search(previous_m)
    previous_month = matchobj.group()
    card100_custom.loc[i,'전월실적'] = previous_month

#5. 할인 = 1, 적립 = 0으로 수정
  if card100_custom.loc[i,'비고'] == '할인':
    card100_custom.loc[i,'비고'] = 1
  else:
    card100_custom.loc[i,'비고'] = 0
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score
```

```python
def get_clf_eval(y_test, pred=None, pred_probas=None):
  confusion = confusion_matrix(y_test, pred)
  accuracy = accuracy_score(y_test, pred)
  precision = precision_score(y_test, pred,average='macro')
  recall = recall_score(y_test, pred,average='macro')
  f1 = f1_score(y_test, pred,average='macro')
  if pred_probas is not None:
    auc_score = roc_auc_score(y_test, pred_probas)
    #print(confusion)
    print(f'정확도 : {accuracy:.4f}, 정밀도 : {precision:.4f}, 재현율 : {recall:.4f}, f1_score:{f1:.4f}, AUC:{auc_score:.4f}')
    return  
  #print(confusion)
  print(f'정확도 : {accuracy:.4f}, 정밀도 : {precision:.4f}, 재현율 : {recall:.4f}, f1_score:{f1:.4f}')
```
